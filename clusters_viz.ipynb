```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gdf = gpd.read_file("/home/libe2152/optimizedsampling/0_data/admin_gdfs/usavars/population/gdf_counties_2015.geojson")
```

```python
cluster_ids = gdf['combined_county_id'].astype(str).values
```

```python
cluster_sizes = []
for id in np.unique(cluster_ids):
    gdf_cluster = gdf[gdf['combined_county_id'] == id]
    cluster_sizes.append(len(gdf_cluster))
```

```python

plt.figure(figsize=(10, 6))
bins = np.arange(0, 10, 1)
plt.hist(cluster_sizes, bins=bins, edgecolor='black')
plt.xlabel('Number of observations per cluster')
plt.ylabel('Frequency')
plt.title('Distribution of Cluster Sizes for USAVars Population')
plt.grid(True, alpha=0.3)
plt.xticks(range(0, 10, 2))
plt.show()
```
