```python
def run_and_plot(alpha, beta, gamma):
    write_cost_to_csv("results/TestSetPerformanceRandomWithCost_formatted.csv", "results/TestSetPerformanceRandomWithCost_edit.csv", "random", alpha, beta, gamma)
    write_cost_to_csv("results/TestSetPerformancelowcostwithCost_formatted.csv", "results/TestSetPerformancelowcostwithCost_edit.csv", "lowcost", alpha, beta, gamma)
    write_cost_to_csv("results/TestSetPerformanceimagewithCost_formatted.csv", "results/TestSetPerformanceimagewithCost_edit.csv", "image", alpha, beta, gamma)

    df_srs = pd.read_csv("results/TestSetPerformanceRandomWithCost_edit.csv", index_col=0)
    df_lowcost = pd.read_csv("results/TestSetPerformancelowcostwithCost_edit.csv", index_col=0)
    df_img = pd.read_csv("results/TestSetPerformanceimagewithCost_edit.csv", index_col=0)

    fig1 = plot_r2_num_samples_with_cost(["srs", "lowcost", "image"], df_srs, df_lowcost, df_img)
    fig2 = plot_r2_cost(["srs", "lowcost", "image"], df_srs, df_lowcost, df_img)
```
