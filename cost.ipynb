```python
import pandas as pd

%load_ext autoreload
%autoreload 2
```

```python
from feasibility import write_cost_to_csv
from plotting import plot_r2_num_samples_with_cost, plot_r2_cost
```

```python
random_path = "results/TestSetPerformanceRandomWithCost_{version}.csv"
lowcost_path = "results/TestSetPerformancelowcostwithCost_{version}.csv"
image_path = "results/TestSetPerformanceimagewithCost_{version}.csv"
satclip_path = "results/TestSetPerformancesatclipwithCost_{version}.csv"

def run_and_plot(alpha, beta, gamma):
    write_cost_to_csv(random_path.format(version="formatted"), random_path.format(version="edit"), "random", alpha, beta, gamma)
    write_cost_to_csv(lowcost_path.format(version="formatted"), lowcost_path.format(version="edit"), "lowcost", alpha, beta, gamma)
    write_cost_to_csv(image_path.format(version="formatted"), image_path.format(version="edit"), "image", alpha, beta, gamma)
    write_cost_to_csv(satclip_path.format(version="formatted"), satclip_path.format(version="edit"), "image", alpha, beta, gamma)

    df_srs = pd.read_csv(random_path.format(version="edit"), index_col=0)
    df_lowcost = pd.read_csv(lowcost_path.format(version="edit"), index_col=0)
    df_img = pd.read_csv(image_path.format(version="edit"), index_col=0)
    df_sat = pd.read_csv(satclip_path.format(version="edit"), index_col=0)

    fig1 = plot_r2_num_samples_with_cost(["srs", "image", "satclip", "lowcost"], df_srs, df_img, df_sat, df_lowcost)
    fig2 = plot_r2_cost(["srs", "image", "satclip", "lowcost"], df_srs, df_img, df_sat, df_lowcost)
```

```python
alpha = 1000
beta = 1
gamma = 4
run_and_plot(alpha, beta, gamma)
```
