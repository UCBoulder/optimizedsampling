```python
from plotting import plot_r2_num_samples, plot_r2_cost
from feasibility import write_cost_to_csv
```

```python
random_path = "results/TestSetPerformanceRandomWithCost_{version}.csv"
lowcost_path = "results/TestSetPerformancelowcostwithCost_{version}.csv"
image_path = "results/TestSetPerformanceimagewithCost_{version}.csv"
satclip_path = "results/TestSetPerformancesatclipwithCost_{version}.csv"

def run_and_plot(alpha, beta, gamma, c):
    write_cost_to_csv(random_path.format(version="formatted"), random_path.format(version="edit"), "random", alpha, beta, gamma, c)
    write_cost_to_csv(image_path.format(version="formatted"), image_path.format(version="edit"), "image", alpha, beta, gamma, c)
    write_cost_to_csv(satclip_path.format(version="formatted"), satclip_path.format(version="edit"), "image", alpha, beta, gamma, c)

    df_srs = pd.read_csv(random_path.format(version="edit"), index_col=0)
    df_img = pd.read_csv(image_path.format(version="edit"), index_col=0)
    df_sat = pd.read_csv(satclip_path.format(version="edit"), index_col=0)

    fig2 = plot_r2_cost(["srs", "img", "satclip"], df_srs, df_img, df_sat)
    return fig2
```

```python
alpha = 1
beta = 100
gamma = 1
c = 0
run_and_plot(alpha, beta, gamma, c).savefig("R^2 vs cost.png")
```
